# Image Processing Code Blocks

```python
from PIL import Image
import os
```

```python
#identify sizes of all images in a directory
img_dir = "../../../Desktop/CAPSTONE/Images/thumbs/medium/"
size_images = dict()
```

```python
for dirpath, _, filenames in os.walk(img_dir):
    for path_image in filenames:
        image = os.path.abspath(os.path.join(dirpath, path_image))
        with Image.open(image) as img:
            width, height = img.size
            size_images[path_image] = (width, height)
print(size_images.values())
```

```python
print(len(size_images))
```

```python
from skimage import io
```

```python
#display images with portrait orientation
for img_item in size_images.items():
    if img_item[1][0] < img_item[1][1]:
        img_file = img_item[0]
        img = io.imread('../../../Desktop/CAPSTONE/Images/thumbs/medium/{}'.format(img_file))
        io.imshow(img)
```
